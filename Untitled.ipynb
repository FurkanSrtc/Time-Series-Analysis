```python

import matplotlib.pyplot as plt
import yfinance as yf
import pandas as pd
import numpy as np
import seaborn as sns
from datetime import datetime, timedelta, timezone

import warnings
warnings.filterwarnings('ignore')

```

```python
stock_list = ['AAPL']
stock_prices = yf.download(stock_list,start = '2018-01-01',end = '2020-01-01',
                              interval='1d', prepost=True)
stock_prices["Average"] = (stock_prices["High"] + stock_prices["Low"])/2
stock_prices.head()
```

```python


plt.figure(figsize=(15, 4))
plt.subplot(121)
sns.lineplot(x = stock_prices.index, y = stock_prices.Volume)
plt.title("Volume & Date")
plt.xlabel("Date")
plt.ylabel("Volume")
plt.xticks(rotation = 45)

plt.subplot(122)
sns.lineplot(x = stock_prices.index, y = stock_prices.Average)
plt.title("Average Prices")
plt.xlabel("Date")
plt.ylabel("Average Price")
plt.xticks(rotation = 45)

plt.show()

```
